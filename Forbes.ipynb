```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("forbes billionaire list cleaned.csv")
df
```

```python
#check missing value
df.isna().sum() 
```

```python
#check value counts of industry
df['industry'].value_counts()
```

```python
#Age kosong digantikan dengan age mean
age=df['age'].mean()
df["age"] = df["age"].replace(np.nan,age)
df
```

```python
#check perubahan age
df.isnull().sum() 
```

```python
df.drop('source',axis=1,inplace=True)
df
```

```python
#Delete every column that have a missing value
df=df.dropna(axis='rows')
df
```

```python
df=df.rename(columns={'country / territory': 'country'})
df=df.rename(columns={'net_worth (in billion $)': 'net'})
df
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x='age', y='net', data=df)
plt.title('Net vs Age')
plt.xlabel('Age')
plt.ylabel('Net (in billions)')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x='industry', y='net', data=df)
plt.title('industry vs net')
plt.xlabel('industry')
plt.ylabel('net')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x='country', y='net', data=df)
plt.title('Coutry vs net')
plt.xlabel('Country')
plt.ylabel('Net (in billions)')
plt.xticks(rotation=90)
plt.show()
```

```python
for n in ('country','industry'):
    plt.figure(figsize=(10, 10))
    sns.countplot(x=n, data=df)
    plt.title(f'diagram tunggal {n}')
    plt.xticks(rotation=90)
    plt.show()
```

```python
negara=df['country'].value_counts().head(10)
plt.figure(figsize=(10, 10))
sns.barplot(x=negara.values, y=negara.index, data=df)
plt.title('Country')
plt.xlabel('Country')
plt.ylabel('Net (in billions)')
plt.xticks(rotation=90)
plt.show()
```
